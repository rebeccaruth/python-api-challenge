```python
#Dependencies
import matplotlib.pyplot as plt
import requests
from scipy import stats
from scipy.stats import linregress
import pandas as pd
import json
import numpy as np
import random

#import citipy
from citipy import citipy

#import keys
from api_keys import weather_api_key
from api_keys import g_key
```

## Generate Cities List

```python
#latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
lat_lng = []

#city list
cities = []

#generate random latitude and longitude
lats = np.random.uniform(low=-90, high=90, size=1500)
lngs = np.random.uniform(low=-180, high=180, size=1500)
lat_lng = zip(lats, lngs)

#find city from latitudes and longitudes
for lat_lng in lat_lng:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    if city not in cities:
        cities.append(city)

#city count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#openweather url
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"
# Build partial query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="

#a list to store data
city_data = []

#before for loop
print("Beginning Data Retrieval     ")
print("-----------------------------")

#for loop
for city in cities:
    city_url = query_url + city
    print(city_url)
    try:
        city_weather = requests.get(city_url).json()
        city_lat = city_weather['coord']['lat']
        city_lng = city_weather['coord']['lon']
        city_mtemp = city_weather['main']['temp_max']
        city_humidity = city_weather['main']['humidity']
        city_cloudiness = city_weather['clouds']['all']
        city_wspeed = city_weather['wind']['speed']
        city_data.append({"City":city,
                      "Latitude":city_lat,
                      "Longitude": city_lng,
                      "Max Temp":city_mtemp,
                      "Humidity":city_humidity,
                      "Cloudiness":city_cloudiness,
                      "Wind Speed":city_wspeed,
                      })
    except:
        print("city not found")
        pass
    
    
#end of for loop
print("--------------------")
print("Data Retrieval Complete")
print("--------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#make city_data into a dataframe
city_df = pd.DataFrame(city_data)
#rename columns
city_df = city_df [["City","Cloudiness","Humidity","Latitude","Longitude","Max Temp","Wind Speed"]]
city_df.head()
```

## Inspect the data and remove the cities where the humidity > 100%.
----
Skip this step if there are no cities that have humidity > 100%. 

```python
#  Get the indices of cities that have humidity over 100%.
city_df.sort_values('Humidity', ascending=False)
```

## Linear Regression

```python
#split data by hemisphere
northern = clean_city_data.loc[clean_city_data["Latitude"] >= 0.0]
northern.reset_index(inplace=True)
southern = clean_city_data.loc[clean_city_data["Latitude"] < 0.0]
southern.reset_index(inplace=True)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = northern['Latitude']
y_values = northern['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temp (F)')
plt.show()
```
